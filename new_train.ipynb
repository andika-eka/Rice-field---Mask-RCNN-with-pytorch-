```python
import os
import json
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
```

```python
#directory
ROOT_DIR  = "./resize_512"
IMAGE_DIR = "./resize_512/image"
ANNOTATION_DIR =  "./resize_512/json"
MODEL_DIR = "./resize_512/result/model"
```

```python
data = []
for  image,annot in zip(os.listdir(IMAGE_DIR), os.listdir(ANNOTATION_DIR)):
    with open(os.path.join(ANNOTATION_DIR,annot), "r") as file:
        dictionary = json.load(file)
        dictionary["file"] = os.path.join(IMAGE_DIR, image)
        data.append(dictionary)

data = pd.DataFrame(data)
```

```python
data_length = len(data)
w,h = 20,20

fig=plt.figure(figsize=(20, 20))
cols , rows = 5, 10

for i in range(cols*rows):
    if (i) == data_length:
        break
    image = plt.imread(data.loc[i,"file"])
    fig.add_subplot(rows, cols, i+1)
    plt.imshow(image)

plt.show()
```
